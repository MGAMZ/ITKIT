Filter the patched data sample generated by `itk_patch` command to prevent too much pure-background patches.

```python
import os, json
from pathlib import Path
from random import random

def parse_patch_meta(folder: Path):
    meta = json.load(open(folder / "crop_meta.json"))["patch_meta"]
    for seriesUID, series_meta in meta.items():
        for patched_sample_name, cls_in_patch in series_meta["class_within_patch"].items():
            if cls_in_patch == [0]:
                yield patched_sample_name

def delete_pure_background_patches(folder: Path, prob_to_still_keep):
    for i, patched_sample_name in enumerate(parse_patch_meta(folder)):
        if random() > prob_to_still_keep:
            os.remove(folder / 'image' / patched_sample_name)
            os.remove(folder / 'label' / patched_sample_name)
            print(f"Deleted {i}: {patched_sample_name}")
```

```python
delete_pure_background_patches(
    folder = Path("/mnt/wsl/Fwsldatavhdx/mgam_datasets/LiTS/spacing1-0.6-0.6/patch80"),
    prob_to_still_keep = 0.1,
)
```
